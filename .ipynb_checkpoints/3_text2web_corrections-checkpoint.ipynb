```python
# EXERCICE 1

def build_fbi_list():
    characters = {
        "Douglas Jones": "travaille dans une compagnie d'assurances",
        "Sarah Palmer": "femme au foyer",
        "Denise Bryson": "membre du FBI",
        "The Fireman": "esprit",
        "Gordon Cole": "membre du Federal Bureau of Investigation",
        "Lucy Moran": "policière",
        "Jerry Horne": "homme d'affaire",
        "Albert Rosenfeld": "membre du Federal Bureau of Investigation",
        "Dick Tremayne": "vendeur",
        "The arm": "esprit",
        "Diane Evans": "membre du FBI",
        "Tammy Preston": "membre du FBI"
    }
    fbi_list = []  # la 1e liste à remplir
    not_fbi_list = []  # la 2e liste à remplir
    for k, v in characters.items():  # on itère sur les clés-valeurs du dico
        # if = 1e manière d'indiquer qu'une personne est membre du FBI
        if v == "membre du FBI":
            fbi_list.append(k)
        # elif = 2e manière d'indiquer qu'une personne n'est pas membre du FBI
        elif v == "membre du Federal Bureau of Investigation":
            fbi_list.append(k)
        # else: si la personne n'est pas membre du FBI.
        else:
            not_fbi_list.append(k)
        
    # quand on a finit de boucler sur characters, on a fini de
    # constituer nos listes. on peut donc faire nos prints
    print("fbi list:", fbi_list)
    print("not fbi list:", not_fbi_list)
    
    # enfin, on retrourne la liste de membres du fbi
    return fbi_list  # la valeur de return est retournée en Out[] par le notebook

# exécuter la fonction
build_fbi_list()

# résultats attendus:
# fbi list: ['Denise Bryson', 'Gordon Cole', 'Albert Rosenfeld', 'Diane Evans', 'Tammy Preston']
# not fbi list: ['Douglas Jones', 'Sarah Palmer', 'The Fireman', 'Lucy Moran', 'Jerry Horne', 'Dick Tremayne', 'The arm']
```

```python
# EXERCICE 4

# on ratrappe le code déjà fait
html = ""  # structure d'accueil pour le html complet

# on récupère la variable source_tolist créée dans l'exercice 2
# on itère sur tous les items pour accéder à chaque item (entrée de catalogue) individuellement
for item in source_tolist:
    item_html = "<div>"  # variable pour stocker l'html de l'item
    list_html = "<ul>"  # variable pour stocker une liste description-prix de chaque item
    
    # on itère sur chaque sous-liste (c-a-d, chaque partie de l'entrée)
    for part in item:  
        # ensuite, il faut déterminer sur quel type d'entrée en travaille,
        # c'est-à-dire repérer un motif présent dans toutes les entrées.
        # dans ce cas ne pas hésiter à jeter un œil au document source.
        # les choses sont bien faites, on a prévu qu'il y ait des motifs
        # dans toutes les parties:
        # - "--" et "~" permettent d'identifier que la "part" est le titre 
        #   (et de faire la séparation entre les différentes parties du titre)
        # - tous les prix sont suivis de la mention "francs."
        # - par élimination, la 3e "part" de chaque "item" est une description
        # on a donc:
        if "--" in part:
            # c'est un titre. il va falloir le séparer en 3 parties:
            # n° d'item, nom et description de la personne
            # on commence par séparer le numéro du reste
            part = part.split("--")
            numero = part[0]
            title = part[1]
            
            # ensuite, on sépare title entre nom et description
            title = title.split("~")
            name = title[0]  # le nom vient en première partie du titre
            # pour extraire la description (trait), on fait les
            # choses de manière élégante: il n'y a pas toujours
            # de description. donc, on choisir de ne rajouter un élément <i>
            # seulement si il y a une description
            if len(title) > 1:
                trait = title[1]
            else:
                trait = ""
                
            # on complète la chaîne item_html avec le titre
            # la syntaxe f"" permet d'insérer des variables (écrites entre
            # {}) à l'intérieur d'une chaîne de caractère
            item_html += f"<dt>Item n°{numero}</dt><dd>{name}: <i>{trait}</i></dd>"
            
        # on passe ensuite aux prix. on considère qu'ils sont tous en francs
        elif "francs." in part:
            price = part
            list_html += f"<li>{price}</li>"
            
        # par élimination, l'item de liste restant est une descrption
        else:
            desc = part
            list_html += f"<li>{desc}</li>"
    
    # on a finit de traiter notre item. on ferme la liste, on l'ajoute à item_html
    list_html += "</ul>"
    item_html += list_html
    item_html += "</div>"  # on ferme notre item
    html += item_html  # on rajoute notre item au html contenant tous les items


# ====================== REPONSE ====================== #
# on ouvre un squelette html a été créé dans utils. on l'ouvre et remplace '{PLACEHOLDER}'
# par le html qu'on vient de produire
with open("utils/catalog_web_skeleton.html", mode="r", encoding="utf-8") as fh:
    skelly = fh.read()
output = skelly.replace("{PLACEHOLDER_BASE}", html)  # on remplace {PLACEHOLDER} par notre html

# on écrit le contenu de output dans un fichier en sortie
with open("output/catalog_web.html", mode="w", encoding="utf-8") as fh:
    fh.write(output)
```

```python
# EXERCICE 5 (BONUS) - exprimer `source` en dictionnaire

# en fait le processus est assez semblable à celui de la
# création d'une liste imbriquée. ce qui change, c'est qu'on assigne
# chaque partie à une clé de dictionnaire

with open("./input/source.txt", mode="r", encoding="utf-8") as fh:
    source = fh.read()

source_todict = {}  # variable d'accueil
source_tolist = []  # on devra passer par une liste
counter = 0  # pour numéroter nos entrées de dictionnaire

source_tolist = source.split(r"\n\n\n")
for item in source_tolist:
    if item == " ":
        continue  # on saute l'élément courant si il est vide
    counter += 1  # on numérote l'item en incrémentant le compteur
    item = item.split(r"\n")  # on crée une liste pour chaque entrée
    for part in item:
        # si c'est un titre
        if "--" in part or "~" in part: 
            part = part.split("--")
            number = part[0]
            
            # exemple de syntaxe rapide: on sépare la 2e partie de part en liste avec le
            # séparateur "~" ; de cette liste, on choisit le 1er élément, à l'index 0
            # et on l'assigne au nom; on choisit le deuxième élément (index 1) et on
            # l'assigne à trait
            if "~" in part[1]:  # si il y a un trait
                name = part[1].split("~")[0]
                trait = part[1].split("~")[1]
            else:
                name = part[1]
                trait = ""
                
            # on ajoute des données à un dictionnaire en faisant dico["clé"] = valeur.
            # si la clé existe déjà, l'ancienne valeur est remplcée par la nouvelle. 
            # pour créer notre clé, on utilise la syntaxe f"{}" pour concaténer "item" 
            # avec le numéro de l'item. ex: item1, item2
            source_todict[f"item{counter}"] = {
                "titre": {
                    "numero": number,
                    "nom": name,
                    "trait": trait   
                }
            }
            
        # si c'est un prix
        elif "francs." in part:
            # on ouvre source_todict à la clé de l'item en cours (f"item{counter}")
            # vu que la valeur de source_todict[f"item{counter}"] est elle-même un dictionnaire,
            # on crée une clé "prix" à ce dictionnaire. on assigne à cette clé
            # la valeur `part`, soit la partie de l'item sur laquelle on itère actuellement
            source_todict[f"item{counter}"]["prix"] = part
            
        # sinon, c'est une description
        else:
            source_todict[f"item{counter}"]["description"] = part
            
print(source_todict)
```

```python
# EXERCICE 6 - À L'AIDE DES VARIABLES CI-DESSOUS, NETTOYER SOURCE

# on ratrappe ce qui a été fait dans l'autre notebook
with open("./input/source.txt", mode="r", encoding="utf-8") as fh:
    source = fh.read()
source = source.lower()

# variables pour faire le remplacement
punct = ['!', '@', '#', '$', '%', '^', '&', '*', '(', ')', '_',
         '+', '=', '{', '}', '[', ']', ':', ';', '"', "'", '|',
         '<', '>', ',', '.', '?', '/', '~', '`']
accents = {"é": "e", "è": "e", "ç": "c",  "à": "a", "ê": "e", "â": "a", "ô": "o",
           "ò": "o", "ï": "i", "ì": "i", "ö": "o"}

# la ponctuation d'abord. on accède à chaque élément de ponctuation individuellement.
# pour chaque élément, on le supprime dans le texte originel. ce qui se passe dans python
# c'est que l'interpréteur, pour chaque élément de ponctuation `p`, lit la variable `source`,
# vérifie si `p` est dans `source`. si oui, il replace `p` par du vide. une fois que il a
# fini le remplacement, il assigne à `source` le texte produit. en d'autres termes,
# à chaque itération, `source` est redéfini avec un signe de ponctuation en moins.
for p in punct:
    source = source.replace(p, "")
    
# ensuite, les accents. avec .items(), le fonctionnement est analogue. on accède 
# au couple clé-valeur (soit caractère_à_remplacer - caractère_de_remplacement) 
# et on effectue le remplacement en remplaçant k par v.
for k, v in accents.items():
    source = source.replace(k, v)
    
print(source)  # on vérifie que tout va bien
```

```python
# EXERCICE 7 - COMPTER LE NOMBRE D'OCCURRENCES DE CHAQUE MOT

sourcelist = source.split()  # on scinde la chaîne en mots et la stocke dans une liste: python
#                              traite la liste item par item, et conserve notre tokenisation
occurrences = {}  # dictionnaire pour stocker les occurrences

for s in sourcelist:  # on accède aux mots
    # on vérifie que le mot `s` remplisse les conditions
    if len(s) > 2 and s != r"\n\n\n":
        # maintenant, le plus difficile: on doit vérifier si le mot est déjà 
        # une clé dans notre dictionnaire. si il l'est déjà, on incrémente la
        # valeur associée par 1; sinon, on crée la clé et on lui assigne la valeur 1.
        # pour vérifier si `s` est une clé, on regarde si `s` fait partie de la liste des clés:
        if s not in occurrences.keys():
             # si ce n'est pas une clé, on crée la clé et lui donne la valeur 1
            occurrences[s] = 1
        else:
            # si c'est une clé, on incrémente sa valeur
            occurrences[s] += 1
            
print(occurrences)
```

```python
# EXERCICE 8 
import json

# on ratrappe ce qui a été fait entre temps
occurrences = {key: value for key, value in sorted(occurrences.items(), key=lambda item: item[1])}
with open("output/occurrences.json", mode="w") as fh:
    json.dump(occurrences, fh, indent=4)  # json.dump(data, fichier, indentation)
with open("output/occurrences.json", mode="r") as fh:
    data = json.load(fh)  # on charge le json dans une variable. c'est un dictionnaire !

    
# ===================== RESULTAT ===================== #
x = list(data.keys())  # l'axe des abscisses correspond à nos clés
y = list(data.values())  # l'axe des ordonnées correspond à nos valeurs
```

```python
# on ratrappe les bouts de code qu'on a fait dans le notebook
from plotly.colors import make_colorscale  # pour le dégradé de couleurs
import plotly.graph_objs as go  # pour faire le graphique
from plotly.io import to_html  # pour écrire sous forme de fichier statique

# pour des raisons purement pratiques, on stocke quelques couleurs dans un dictionnaire
colors = {"white": "#ffffff", "cream": "#fcf8f7", "blue": "#0000ef", "purple": "#bd148b"}
scale = make_colorscale([colors["blue"], colors["purple"]])  # créer une échelle de couleurs
colorbar = {"xpad": 0, "ypad": 0}  # pour un joli dégradé qui s'affiche sur la droite

fig = go.Figure(
    data=[go.Bar(x=x, y=y, width=1.1, marker={"color": y, "colorbar":colorbar,
                                              "colorscale": scale, "cauto": False, 
                                              "cmin": 0, "cmax": 10},
                 hovertemplate="<b>'%{x}'</b>: %{y} occurrences<extra></extra>")],
    layout=go.Layout({
        "paper_bgcolor": colors["white"],
        "plot_bgcolor": colors["cream"],
        "margin": {"l": 5, "r": 5, "t": 30, "b": 30},
        "yaxis_range": [0, 20],
        "xaxis": {"anchor": "x", "title": {"text": "Mot"}},
        "yaxis": {"anchor": "y", "title": {"text": "Nombre d'occurrences"}},
        "title": "Nombre d'occurrences par mot",
    })
)

htmlfig = to_html(fig, include_plotlyjs="cdn", full_html=False, default_width="100%",
                  default_height="70%")

with open("output/catalog_web.html", mode="r") as fh:
    html = fh.read()  # on lit le fichier source
html = html.replace("{PLACEHOLDER_GRAPH}", htmlfig)  # on remplace le placeholder par la figure

with open("output/catalog_web.html", mode="w") as fh:
    fh.write(html)
```
